# AVL

---

## How to Calculate the Balance Factor

the standard formula is:
- `Balance Factor = Height(Left Subtree) - Height(Right Subtree)`
    - **Height**: The number of *edges* on the longest downward path from the node to a leaf. (A leaf node itself has a height of 0, and an empty/null child has a height of -1).
    - **Balanced**: An AVL tree is considered balanced if every node has a balance factor of -1, 0, or +1.

---


## How to Perform an LL Rotation

To fix an `LL` imbalance, we perform a `single Right Rotation` on the unbalanced node.
- The middle node is pulled up to become the new root.
- The unbalanced node is pushed down and to the right, becoming the right child of the new root.
- The bottom node remains attached as the left child, which is the new root.

---

### How to Perform an RR Rotation

To fix an `RR` imbalance, we perform a single `Left Rotation` on the unbalanced node.

- The middle node is pulled up to become the new root.
- The unbalanced node is pushed down and to the left, becoming the left child, aka new root.
- The bottom node remains attached as the right child, aka new root.
