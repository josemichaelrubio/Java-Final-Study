# Hash Table Notes

*Hash table* = key based data structure, which allows a decrease in time complexity

*hash function* = takes a key as input and provides an array index into which we store a reference to the key's associated value.

## Collisions

A collision happens when the hash function maps two different keys to the exact same index

### Collision Resolution

*Chaining* = Stroing a list (or a chain) of elements at each index so multiple items can live at the same address

*Closed Hashing (Open Addressing)* = Storeing all elements directly inside the array iteself. IF a spot is taken, the algorithm probes for the next available slot

- Linear Probing = Simply checking the very next slot sequentially using `f(i) = i` . This is easy but can lead to Clustering, where chunks of data bunch up and slow down search times.
- Quadratic Probing: Jumping further away using `f(i) = i^2` to reduce that clustering effect.
- Double Hashing: Using a completely separate, second hash function (like `h_2(x) = 7 - (x % 7)`) to determine the jump size, spreading data out even more evenly.

```java
public class LinearProbingHashTable {
    private Integer[] table;
    private int size;

    // Constructor to initialize the table
    public LinearProbingHashTable(int size){
        this.size = size;
        this.table = new Integer[size]; // All slots default to 'null'
    }

    // Hash Function: h(k) = k % size
    private int hash(int key){
        return key % size;
    }

    // Insert method using Linear Probing
    public void insert(int key){
        int index = hash(key);
        int probes = 1; // Start with the initial check

        // Linear Probing step: 
        // Keep moving forward as long as the current slot is NOT empty
        while (table [index] != null){
            index = (index + 1) % size;
            probes++;
        }

        // We found a null (empty) slot! Place the key here.
        table[index] = key;
        System.out.println("Inserted " + key + " at index " + index + " (Took " + probes + " probes)");
    } 

    // Helper method to print the table exactly like your notes
    public void printTable() {
        System.out.print("Final Table: [");
        for (int i = 0; i < size; i++) {
            System.out.print((table[i] == null ? "x" : table[i]));
            if (i < size - 1) System.out.print(", ");
        }
        System.out.println("]");
    }

    public static void main(String[] args) {
        LinearProbingHashTable ht = new LinearProbingHashTable(10);
        int[] keysToInsert = {12, 18, 13, 2, 3, 23, 5, 15};

        for (int key : keysToInsert) {
            ht.insert(key);
        }
        
        System.out.println("-------------------------");
        ht.printTable();
    }
}

// For In IJava notebooks, you need call main explicitly
LinearProbingHashTable.main(null);
```
