# Hash Table Notes

*Hash table* = key based data structure, which allows a decrease in time complexity

*hash function* = takes a key as input and provides an array index into which we store a reference to the key's associated value.

## Collisions

A collision happens when the hash function maps two different keys to the exact same index

### Collision Resolution

*Chaining* = Stroing a list (or a chain) of elements at each index so multiple items can live at the same address

*Closed Hashing (Open Addressing)* = Storeing all elements directly inside the array iteself. IF a spot is taken, the algorithm probes for the next available slot

- Linear Probing = Simply checking the very next slot sequentially using `f(i) = i` . This is easy but can lead to Clustering, where chunks of data bunch up and slow down search times.
- Quadratic Probing: Jumping further away using `f(i) = i^2` to reduce that clustering effect.
- Double Hashing: Using a completely separate, second hash function (like `h_2(x) = 7 - (x % 7)`) to determine the jump size, spreading data out even more evenly.
